```python
from simworld.utils.video_recorder import VideoRecorder
from simworld.utils.vector import Vector
from simworld.communicator.communicator import Communicator
from simworld.communicator.unrealcv import UnrealCV
from simworld.agent.humanoid import Humanoid
```

```python
humanoid = Humanoid(Vector(600, 9000), Vector(1, 0)) # position here has default z-axis value 600
```

```python
communicator = Communicator(UnrealCV(port=9100, ip='132.239.95.132'))
communicator.spawn_agent(humanoid, '', position=[600, 9000, 600]) # reset humanoid position here to overwrite z-axis position
communicator.humanoid_set_speed(humanoid.id, 100)
```

```python
# print(communicator.unrealcv.client.request('vget /cameras'))
image = communicator.get_camera_observation(1, viewmode='lit')

print(image.shape)

import cv2
import numpy as np

cv2.imwrite('test.png', cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
```

```python
def move_pattern(step):
    if step < 25:
        communicator.humanoid_set_speed(humanoid.id, 100)
    else:
        communicator.humanoid_set_speed(humanoid.id, 0)

video_recorder = VideoRecorder(
    communicator,
    humanoid,
    output_dir='./video_record',
    video_name='output.mp4',
    resolution=(1440, 720),
    fps=25.0,
    frame_num=50,
    move_pattern=move_pattern,
)
```

```python
video_recorder.record()
```
