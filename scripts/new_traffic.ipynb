```python
import sys
from pathlib import Path
sys.path.append(str(Path().resolve().parent))
from simworld.traffic.base.traffic_signal import TrafficSignal
from simworld.communicator.unrealcv import UnrealCV
from simworld.communicator.communicator import Communicator
from simworld.config import Config
from simworld.utils.vector import Vector
```

```python
config = Config()
```

```python
unrealcv = UnrealCV()
communicator = Communicator(unrealcv)
```

```python
signal_1 = TrafficSignal(Vector(500, 500), Vector(-1, 0), 0, 0, 'both')
signal_2 = TrafficSignal(Vector(500, -500), Vector(0, 1), 0, 0, 'both')
signal_3 = TrafficSignal(Vector(-500, 500), Vector(0, -1), 0, 0, 'both')
signal_4 = TrafficSignal(Vector(-500, -500), Vector(1, 0), 0, 0, 'both')
signals = [signal_1, signal_2, signal_3, signal_4]
```

```python
communicator.spawn_traffic_signals(signals, config['traffic.traffic_signal.traffic_light_model_path'], config['traffic.traffic_signal.pedestrian_light_model_path'])
```

```python
communicator.spawn_intersection('BP_Intersection_1', '/Game/TrafficSystem/Intersection/BP_Intersection.BP_Intersection_C')
```

```python
for signal in signals:
    communicator.add_traffic_signal('BP_Intersection_1', signal)
```

```python
communicator.traffic_signal_start_simulation('BP_Intersection_1')
```
